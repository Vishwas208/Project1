```python
global all_std
all_std=[]
def createnewdict():
    new_std={'Name':'','Roll no.':0,'Marks1':0,'Marks2':0 }
    while True:
        Name=input('Enter Name: ')
        new_std['Name']=Name
        try:
            Rollno=int(input('Enter the roll number: '))
        except ValueError:
            print('Enter a valid number')
            continue
        else:
            new_std['Roll no.']=Rollno
        try:
            Marks1=int(input('Enter marks of subject 1: '))
        except ValueError:
            print('Enter a valid number!')
            continue
        else:
            new_std['Marks1']=Marks1
        try:
            Marks2=int(input('Enter marks of subject 2: '))
        except ValueError:
            print('Enter a valid number!')
            continue
        else:
            new_std['Marks2']=Marks2
        return new_std
def display():
    if not all_std:
            print('No student profile created yet!')
    else:
        for i,j in enumerate(all_std):
            print(i,j)
def delete(a):
    for i in all_std:
        if i['Roll no.']==a:
            print(i)
            all_std.pop(all_std.index(i))
            print('This student entry has been deleted')
        else:
            print('Roll no. not found!')
def search(b):
    for i in all_std:
        if i['Roll no.']==b:
            print('The required student entry is:-')
            print(i)
            
        

def main():
    while True:
        try:
            action=int(input('''
            Enter 1: To create a new Student profile
            Enter 2: To display all student profile
            Enter 3: To delete a specific Student profile
            Enter 4: To search a Student profile
            Enter 5: To exit the program
            '''))
        except ValueError:
            print('\n Enter a valid number!!!')
        else:
            print('\n')
        if action==1:
            new_std=createnewdict()
            all_std.append(new_std)
        elif action==2:
            display()
        elif action==3:
            x=int(input("enter the Roll no of the student"))
            delete(x)
        elif action==4:
            y=int(input('Enter the Roll no. to search'))
            search(y)
        elif action==5:
            break
        else:
            print('Invalid input. Please enter a valid input')

if __name__ == "__main__":
    main()
        
        
    
    
```
